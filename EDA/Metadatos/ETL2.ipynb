**Segmentacion del df_final por palabras claves relacionadas con restaurantes y comidas, objeto del analisis y se obtiene un nuevo df_final_1, segmentado solo con la categoria: Restaurantes y afines**

**Remplazo valores nulos por ('')**

```python
df_final = df_final.fillna('')
```

**Segmentacion por categoria de negocio**

```python
# Palabras clave relacionadas con la comida y los restaurantes
keywords = ['restaurant', 'food','grill', 'cafe', 'pizzeria', 'bakery','breakfast', 'coffee shop', 'hamburger', 'sandwich shop']

# Función para verificar si alguna palabra clave está en la lista de categorías
def is_food_related(categories):
    return any(keyword in category.lower() for category in categories for keyword in keywords)

# Aplica la función a la columna 'category' para crear una nueva columna 'is_food_related'
df_final['is_food_related'] = df_final['category'].apply(is_food_related)

# Filtra el DataFrame para incluir solo las filas donde 'is_food_related' es True
df_final_1 = df_final[df_final['is_food_related']]

# Muestra el nuevo DataFrame
df_final_1
```
